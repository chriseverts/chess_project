```python
import pandas as pd
# ^^^ pyforest auto-imports - don't write above this line
pd.set_option("display.max_rows", None, "display.max_columns", None)
pd.reset_option("display.max_rows", "display.max_columns")
```

```python
# reading csv files
data1 = pd.read_csv('chess_players.csv')
data2 = pd.read_csv('int_chess_stats.csv')
  
# using merge function by setting how='right'. Country is the primary key
df = pd.merge(data1, data2,
                   on='Country',
                   how='left')
  
df
```

```python
# Set index to Country
df = df.set_index('Country')
df
```

```python
df.info()
```

```python
# dropping unamed and flag columns that contain no data or nan
df.drop([col for col in df.columns if "Unnamed" in col], axis=1, inplace=True)
df.drop([col for col in df.columns if "Flag" in col], axis=1, inplace=True)
```
